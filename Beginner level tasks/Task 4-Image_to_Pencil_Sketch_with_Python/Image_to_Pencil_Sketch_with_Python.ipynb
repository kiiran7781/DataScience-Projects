```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
```

```python
def to_pencil_sketch(image):

    gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    inverted_image = cv2.bitwise_not(gray_image)
    blurred_image = cv2.GaussianBlur(inverted_image, (115, 115), sigmaX=0, sigmaY=0)
    inverted_blurred_image = cv2.bitwise_not(blurred_image)
    pencil_sketch = cv2.divide(gray_image, inverted_blurred_image, scale=256.0)
    desired_width = 520
    desired_height = 400
    resized_pencil_sketch = cv2.resize(pencil_sketch, (desired_width, desired_height))

    return resized_pencil_sketch
```

```python
input_img = cv2.imread('/content/360_F_177476718_VWfYMWCzK32bfPI308wZljGHvAUYSJcn.jpg')

if input_img is None:
    print("Error: Image not loaded.")
else:
    pencil_sk = to_pencil_sketch(input_img)
    print(cv2_imshow(input_img),cv2_imshow(pencil_sk))


```
